```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL']='3' #로그를 담당하는 TensorFlow 환경변수
                                       #1=INFO, 2=WARNING, 3=ERROR 로그를 filter한다.
import tensorflow as tf
import numpy as np

#[털, 날개]
x_data = np.array([[0, 0], [1, 0], [1, 1], [0, 0], [0, 0], [0, 1]])
y_data = np.array([
    [1, 0, 0],  # 기타
    [0, 1, 0],  # 포유류
    [0, 0, 1],  # 조류
    [1, 0, 0],
    [1, 0, 0],
    [0, 0, 1]
])

X = tf.placeholder(tf.float32, name="X") #Constant, Variable, Placeholder
Y = tf.placeholder(tf.float32, name="Y")

W = tf.Variable(tf.random_uniform([2, 3], -1, 1))
b = tf.Variable(tf.zeros([3]))
L = tf.add(tf.matmul(X, W), b)
L = tf.nn.relu(L, name="L")
model = tf.nn.softmax(L)
cost = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(model), axis=1))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train_op = optimizer.minimize(cost)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    merged = tf.summary.merge_all()
    #writer = tf.summary.FileWriter("./logs", sess.graph)

    for step in range(100):
        sess.run(train_op, feed_dict={X: x_data, Y: y_data})
        if(step + 1) % 10 == 0:
            print(step + 1, sess.run(cost, feed_dict={X: x_data, Y: y_data}))

    #########
    # 결과 확인
    # 0: 기타 1: 포유류, 2: 조류
    ######
    # tf.argmax: 예측값과 실제값의 행렬에서 tf.argmax 를 이용해 가장 큰 값을 가져옵니다.
    # 예) [[0 1 0] [1 0 0]] -> [1 0]
    #    [[0.2 0.7 0.1] [0.9 0.1 0.]] -> [1 0]
    prediction = tf.argmax(model, 1)
    target = tf.argmax(Y, 1)
    print('예측값:', sess.run(prediction, feed_dict={X: x_data}))
    print('실제값:', sess.run(target, feed_dict={Y: y_data}))

    is_correct = tf.equal(prediction, target)
    accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))
    print('정확도: %.2f' % sess.run(accuracy * 100, feed_dict={X: x_data, Y: y_data}))
```

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL']='3'
import tensorflow as tf
import numpy as np

#[털, 날개]
x_data = np.array([[0, 0], [1, 0], [1, 1], [0, 0], [0, 0], [0, 1]])
y_data = np.array([
    [1, 0, 0],  # 기타
    [0, 1, 0],  # 포유류
    [0, 0, 1],  # 조류
    [1, 0, 0],
    [1, 0, 0],
    [0, 0, 1]
])

# print(x_data)
# print(y_data)

X = tf.placeholder(tf.float32, name="X")
Y = tf.placeholder(tf.float32, name="Y")

W1 = tf.Variable(tf.random_uniform([2, 10], -1, 1))
W2 = tf.Variable(tf.random_uniform([10, 3], -1, 1))
b1 = tf.Variable(tf.zeros([10]))
b2 = tf.Variable(tf.zeros([3]))


L1 = tf.add(tf.matmul(X, W1), b1)
L1 = tf.nn.relu(L1, name="L1")
model = tf.add(tf.matmul(L1, W2), b2)
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=Y, logits=model))

optimizer = tf.train.AdamOptimizer(learning_rate=0.01)
train_op = optimizer.minimize(cost)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    merged = tf.summary.merge_all()
    #writer = tf.summary.FileWriter("./logs", sess.graph)

    for step in range(100):
        sess.run(train_op, feed_dict={X: x_data, Y: y_data})
        if(step + 1) % 10 == 0:
            print(step + 1, sess.run(cost, feed_dict={X: x_data, Y: y_data}))

    #########
    # 결과 확인
    # 0: 기타 1: 포유류, 2: 조류
    ######
    # tf.argmax: 예측값과 실제값의 행렬에서 tf.argmax 를 이용해 가장 큰 값을 가져옵니다.
    # 예) [[0 1 0] [1 0 0]] -> [1 0]
    #    [[0.2 0.7 0.1] [0.9 0.1 0.]] -> [1 0]
    prediction = tf.argmax(model, 1)
    target = tf.argmax(Y, 1)
    print('예측값:', sess.run(prediction, feed_dict={X: x_data}))
    print('실제값:', sess.run(target, feed_dict={Y: y_data}))

    is_correct = tf.equal(prediction, target)
    accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))
    print('정확도: %.2f' % sess.run(accuracy * 100, feed_dict={X: x_data, Y: y_data}))
```
